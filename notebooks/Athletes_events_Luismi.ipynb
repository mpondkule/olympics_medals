```python
df = pd.read_csv("../data/raw/athlete_events.csv")
df.info()
```

```python
df = df.drop(columns=['Games'])
```

```python
df_countries = pd.read_csv('../data/raw/noc_regions.csv')
df_countries.info()
```

```python
df_countries = pd.read_csv('../data/raw/noc_regions.csv')
df_countries.info()
df_countries['NOC'].unique()
df_countries['region'].unique()
dic = {}
for i, k in zip(df_countries['NOC'], df_countries['region']):
    dic[i] = k
dic

```

```python
df['NOC'] = df['NOC'].replace(dic)

df['NOC'].unique()
```

```python
df_1 = df[(df['Year'] >= 1946) & (df['Year'] <= 1991)]
df_1_sp = df[(df['Year'] >= 1946) & (df['Year'] <= 1959)]
df_2_sp = df[(df['Year'] >= 1960) & (df['Year'] <= 1991)]
```
